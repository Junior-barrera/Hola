```python
# Operaciones Básicas
a = 10
b = 3
```

```python
# Suma
suma = a + b
print("Suma:", suma)

# Resta
resta = a - b
print("Resta:", resta)

# Multiplicación
multiplicacion = a * b
print("Multiplicación:", multiplicacion)

# División
division = a / b  # Resultado como número decimal
print("División:", division)
```

```python
# Condicionales if
if a > b:
    print(f"{a} es mayor que {b}")
else:
    print(f"{b} es mayor o igual que {a}")

#  Ciclos for
for i in range(5):  # Repite 5 veces (de 0 a 4)
    print("Iteración:", i)

# Ciclo while
contador = 0
while contador < 3:  # Se repite mientras la condición sea verdadera
    print("Contador:", contador)
    contador = contador + 1 #o contador = contador + 1
```
